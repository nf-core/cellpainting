```python
import pandas as pd
```

```python
data = pd.read_csv("/Users/rachelbozadjian/Desktop/load_data.csv")

data.columns
```

```python
filtered_data = data[["URL_OrigBrightfield_L", "URL_OrigBrightfield_H", "URL_OrigBrightfield", "URL_OrigRNA", "URL_OrigMito", "URL_OrigER", "URL_OrigDNA", "URL_OrigAGP", "Metadata_Source", "Metadata_Batch", "Metadata_Plate", "Metadata_Well", "Metadata_Site", "Metadata_Col", "Metadata_FieldID", "Metadata_PlaneID", "Metadata_Row"]]
filtered_data.head()
```

```python
longer_data = pd.melt(
    filtered_data,
    id_vars=['Metadata_Source', 'Metadata_Batch', 'Metadata_Plate', 'Metadata_Well', 
             'Metadata_Site', 'Metadata_Col', 'Metadata_FieldID', 'Metadata_PlaneID', 'Metadata_Row'],
    value_vars=['URL_OrigBrightfield_L', 'URL_OrigBrightfield_H', 'URL_OrigBrightfield', 
                'URL_OrigRNA', 'URL_OrigMito', 'URL_OrigER', 'URL_OrigDNA', 'URL_OrigAGP'],
    var_name='channel',
    value_name='path'
)

# Clean up the URL_type column to remove the 'URL_Orig' prefix
longer_data['channel'] = longer_data['channel'].str.replace('URL_Orig', '')

longer_data = longer_data.rename(columns={"Metadata_Source":"source", "Metadata_Batch":"batch", "Metadata_Plate":"plate", "Metadata_Well":"well", "Metadata_Site":"site", "Metadata_Col":"col", "Metadata_FieldID":"field_id", "Metadata_PlaneID":"plane_id", "Metadata_Row":"row"})

df_reordered = longer_data[['channel', 'path', 'source', 'batch', 'plate', 'well', 'site', 'col', 'field_id',
       'plane_id', 'row']]

df_reordered
```

```python
pd.DataFrame.to_csv(df_reordered, "samplesheet_test_full.csv", index=False)
```
